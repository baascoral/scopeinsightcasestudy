```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import scipy as sp
from scipy import stats
```

```python
df2 = df.drop_duplicates(subset=["producing_organization_id"], keep="first")
df2
```

```python
df2['perc_ex_male'] = (df2.number_of_male_executives/df2.number_of_executives)*100
df2['perc_ex_female'] = (df2.number_of_female_executives/df2.number_of_executives)*100
df2
```

```python
top3male = df2[df2.perc_ex_male >= 60]['Total score'].nlargest(n=3)
df2.loc[top3male.index]
```

```python
top3female = df2[df2.perc_ex_female >= 60]['Total score'].nlargest(n=3)
df2.loc[top3female.index]
```

```python
#To retrive the products (df) to the Top 3 executive male Coops
top3male_org = df2.loc[top3male.index].producing_organization_id.to_list()
df_top3_male = df[df["producing_organization_id"].isin(top3male_org)]
df_top3_male
```

```python
top3female_org = df2.loc[top3female.index].producing_organization_id.to_list()
top3female_org
df_top3_female = df[df["producing_organization_id"].isin(top3female_org)]
df_top3_female
```

```python

df_p = pd.melt(dimensions, id_vars=['producing_organization_id'], value_vars=['number_of_executives','number_of_members',\
        'Section score'], ignore_index= True)
df_p.corr()
```

```python
vis3 = sns.lmplot(data = dimensions, x = "number_of_executives", y = "number_of_members", fit_reg = True, size = 6, aspect = 1)

def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data['number_of_executives'], data['number_of_members'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p), transform=ax.transAxes)
    
vis3.map_dataframe(annotate)
plt.show()
```
