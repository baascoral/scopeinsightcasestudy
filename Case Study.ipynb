```python
# df3 Drop duplicates per id and section title

df3 = df.drop_duplicates(subset=["producing_organization_id","section_title"], keep="first")
#df3[df3.producing_organization_id == 4677 ]
df3
```

```python
# In df3 filter only the 8 dimensions SCOPEinsight (section_title) 

dimensions_scope = df3.loc[df3['section_title'].isin(["Internal Management", "Financial Management", "Sustainability",\
                    "Operations", "Production Base", "Market", "External Risk", "Enabling Environmental"])]

dimensions = dimensions_scope[["producing_organization_id", "number_of_members", "number_of_executives", "section_title", "Section score"]]
dimensions[dimensions.producing_organization_id == 4779]
```

```python
# Correlation Number of executives Vs Section Score 

df_nex = dimensions.pivot_table(index = ["Section score"],  values = ["number_of_executives"], columns = ["section_title"])

df_nex.corr()
```

```python
# Facet Grid "number_of_executives", "Section score"
g = sns.FacetGrid(dimensions, row ="section_title", hue="section_title")
g = g.map(plt.scatter, "number_of_executives", "Section score")

def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data['number_of_executives'], data['Section score'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p), transform=ax.transAxes)
    
g.map_dataframe(annotate)

plt.show()
```

```python
vis3 = sns.lmplot(data = dimensions, x = "number_of_executives", y = "Section score", hue = "section_title" , fit_reg = True, size = 6, aspect = 1)

def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data['number_of_executives'], data['Section score'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p), transform=ax.transAxes)
    
vis3.map_dataframe(annotate)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
vis3.set(ylim=(0,5))


plt.show()
```

```python
# Correlation Number of Members Vs Section Score 

df_nm = dimensions.pivot_table(index = ["Section score"],  values = ["number_of_members"], columns = ["section_title"])

df_nm.corr()
```

```python
# Facet Grid "number_of_members", "Section score"
g1 = sns.FacetGrid(dimensions, row ="section_title", hue="section_title")
g1 = g1.map(plt.scatter, "number_of_members", "Section score")

def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data['number_of_members'], data['Section score'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p), transform=ax.transAxes)
    
g1.map_dataframe(annotate)

plt.show()
```

```python
#4 Identify the top 5 crops

df_cote_divoire = df[df.country == "Côte d'Ivoire"]

#one id has varios products and each product was evaluated 8 times - drop duplicates per id and product

df_cd = df_cote_divoire.drop_duplicates(subset=["producing_organization_id", "product"], keep="first")
df_cd

#df_cd[df_cd.producing_organization_id == 4479] each id has more than one product. There is a number of member per 
#id but not per product
```

```python
#Group by product and sum the number of members

top5crops = df_cd.groupby('product')['number_of_members'].sum().nlargest(n=5)
top5crops
```

```python
# Filtering the top 5 crops in the df_cd (Cote d'Ivoire) and include the Total Score 

df_crops = df_cd.loc[df_cd['product'].isin(["Cocoa", "Cassava", "Other nuts", "Rice, paddy", "Chillies and peppers"])]
df_crops_fil = df_crops[["product", "number_of_members","Total score"]]
df_crops_fil.sort_values(by=['product'])

```

```python
# Visualize 5 Top Crops Vs. Total Score - Cote d'Ivoire

vis4 = sns.lmplot(data = df_crops_fil, x = "number_of_members", y = "Total score", hue = "product", fit_reg = False)

plt.title("Top 5 Crops Vs. Total Score - Cote d'Ivoire", fontsize=15, color = "Black")
plt.ylabel("Total Score", fontsize=15)
plt.xlabel("Number of Members", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
vis2.set(ylim=(0,5))
```
